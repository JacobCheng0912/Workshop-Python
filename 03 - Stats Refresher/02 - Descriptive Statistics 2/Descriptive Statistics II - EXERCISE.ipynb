### Import Numpy, pandas, and matplotlib

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt 

plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (10, 6)

print('Libraries imported!')
```

### Import the data in the CSV file
Today you'll be working with the tips dataset from the lecture. You can find the data [here](https://raw.githubusercontent.com/axrd/datasets/master/tipdata.csv). 

```python
# Run this cell to import the data

df = pd.read_csv('https://raw.githubusercontent.com/axrd/datasets/master/tipdata.csv', index_col=0)
```

### Print out the first 10 rows

```python
# Display the first 10 rows of the DataFrame
print(df.head(10))
```

### Print out the last 10 rows

```python
# Display the last 10 rows of the DataFrame
print(df.tail(10))
```

### Examine total bill and tip to identify any correlation. Calculate the correlation coefficient, AND plot both features as a scatterplot.

```python
# Calculate the correlation coefficient between "total_bill" and "tip"
correlation_total_bill_tip = df['total_bill'].corr(df['tip'])
print(f'Correlation between total bill and tip: {correlation_total_bill_tip}')
```

```python
# Create a scatterplot of "total_bill" and "tip"
plt.scatter(df['total_bill'], df['tip'], alpha=0.7)
plt.title('Scatterplot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Examine size and total bill to identify any correlation. Calculate the correlation coefficient, AND plot both features as a scatterplot.

```python
# Calculate the correlation coefficient between "size" and "total_bill"
correlation_size_total_bill = df['size'].corr(df['total_bill'])
print(f'Correlation between size and total bill: {correlation_size_total_bill}')
```

```python
# Create a scatterplot of "size" and "total_bill"
plt.scatter(df['size'], df['total_bill'], alpha=0.7)
plt.title('Scatterplot of Size vs Total Bill')
plt.xlabel('Size')
plt.ylabel('Total Bill')
plt.show()
```

### Examine size and tip to identify any correlation. Calculate the correlation coefficient, AND plot both features as a scatterplot.

```python
# Calculate the correlation coefficient between "size" and "tip"
correlation_size_tip = df['size'].corr(df['tip'])
print(f'Correlation between size and tip: {correlation_size_tip}')
```
